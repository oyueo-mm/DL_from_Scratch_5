```python
import numpy as np
import matplotlib.pyplot as plt

# 실험 설정
num_experiments = 10000  # 실험 횟수
n_values = [1, 5, 10, 30]  # 각 실험에서 동전 던지는 횟수 (표본 크기)

plt.figure(figsize=(10, 6))

for n in n_values:
    sample_means = [np.mean(np.random.choice([0, 1], size=n)) for _ in range(num_experiments)]
    plt.hist(sample_means, bins=30, alpha=0.6, label=f'n={n}', density=True)

"""
plt.title("중심극한정리 시뮬레이션 (동전 던지기)")
plt.xlabel("표본 평균")
plt.ylabel("밀도")
"""
plt.legend()
plt.show()

n = 100000
sample_means = [np.mean(np.random.choice([0, 1], size=n)) for _ in range(num_experiments)]
plt.hist(sample_means, bins=30, alpha=0.6, label=f'n={n}', density=True)
plt.legend()
plt.show()
```
